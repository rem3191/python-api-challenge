```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"
#print(url)

max_temps=[]
humidities=[]
lats=[]
lngs=[]
cloudiness=[]
countries=[]
wind_speeds=[]
dates=[]
city_names=[]


for i,city in enumerate(cities):
    query_url=url+"appid="+weather_api_key+"&q="+city
    response=requests.get(query_url).json()
    
    try:
        cloudiness.append(response["clouds"]["all"])
        countries.append(response["sys"]["country"])
        dates.append(response["dt"])
        max_temps.append(response["main"]['temp_max'])
        humidities.append(response["main"]['humidity'])
        lats.append(response["coord"]["lat"])
        lngs.append(response["coord"]["lon"])
        wind_speeds.append(response["wind"]["speed"])
        city_names.append(response["name"])
        print(f"Processing record {i}|{city}")
        
    except (KeyError):
            print("City Not Found. Skipping Entry.")    

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict={
                'City':city_names,
                'Country':countries,
                'Lat':lats,
                'Lng':lngs,
                'Cloudiness':cloudiness,
                'Wind Speed':wind_speeds,
                'Humidity':humidities,
                'Maximum Temp':max_temps,
                'Date':dates
}
city_df=pd.DataFrame(weather_dict)
city_df
```

```python
city_df.to_csv("../output_data/city.csv")
```

## Linear Regression

```python
north_hemp=city_df.loc[city_df["Lat"]>=0]
south_hemp=city_df.loc[city_df["Lat"]<0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.figure(figsize=(10,5))
x_values=north_hemp['Lat']
y_values=north_hemp['Maximum Temp']

(slope,intercept,rvalue,pvalue,stderr)=linregress(x_values, y_values)

regress_values=x_values*slope+intercept

line_eq='y='+str(round(slope,2))+'x+'+str(round(intercept,2))

plt.scatter(x_values,y_values,marker='o',color='red',s=[60],edgecolors='black')
plt.plot(x_values,regress_values,'darkblue',linewidth=3)

plt.title('Northern Hempisphere - Max. Temp Vs. Latitude', fontsize=10)
plt.ylabel('Max Temp (F)',fontsize=10,color='black')
plt.xlabel('Latitude',fontsize=10,color='black')
plt.annotate(line_eq,(1,1),fontsize=15,color='darkblue')

plt.savefig("../Images/Northern Hemisphere - Max Temp Vs. Latitude.png")
plt.show()
print("There appears to be a strong correlation between Maximum Temp and Latitude")
```
